**<h1 style = "color:red;">Question 5: Remove Duplicates</h1>**

**In this step, duplicate rows are identified and removed from the dataset.**

```python
csv_1.duplicated().sum()
```

```python
csv_1.drop_duplicates(inplace=True)
print("Duplicates removed successfully!")
```

**<h1 style ="color:Brown;">Question 6: Create a Derived Column (total_amount = price * quantity)</h1>**

**In this step, a new column named total_amount is created using price and quantity columns.**



```python
#converting discount column into numeric datatype
# Required for mathematical calculations
csv_1['discount'] = pd.to_numeric(csv_1['discount'],errors='coerce')
```

```python
# Replacing missing discount values with 0
# Prevents calculation errors

csv_1['discount'] = csv_1['discount'].fillna(0)
```

```python
# Creating price, quantity, and total_amount columns

csv_1['price'] = csv_1['initial_price'] - csv_1['discount']

csv_1['quantity'] = csv_1['ratings_count']

csv_1['total_amount'] = csv_1['price'] * csv_1['quantity']
```

```python
#Displaying derived columns

csv_1[['price','quantity','total_amount']].head()
```
