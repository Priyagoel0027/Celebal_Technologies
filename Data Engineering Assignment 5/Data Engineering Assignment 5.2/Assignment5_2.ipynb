Spark keeps data in RAM (memory), allowing it to reuse the same data across multiple iterations.

This significantly reduces disk reads and writes, which improves performance.

## Benefits

- Faster execution
- Less disk I/O
- Better support for machine learning algorithms
- Efficient iterative processing

### Example
Algorithms such as K-Means and Linear Regression require multiple iterations. Spark keeps data in memory throughout these iterations, reducing execution time.

Spark speeds up iterative machine learning by keeping data in memory instead of repeatedly accessing disk storage.

## Q4: Given a DataFrame df_sales, write a query to filter for rows where the region is 'West' and then group by product_category to find the average sale_amount.

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg

# Sample Data
data = [
    ("West", "Electronics", 1000),
    ("West", "Electronics", 2000),
    ("West", "Clothing", 500),
    ("East", "Electronics", 1500)
]

columns = ["region", "product_category", "sale_amount"]

# Create DataFrame
df_sales = spark.createDataFrame(data, columns)

# Filter West region
west_sales = df_sales.filter(df_sales.region == "West")

# Group by product_category and find average sale amount
result = west_sales.groupBy("product_category") \
                   .agg(avg("sale_amount").alias("avg_sale_amount"))

result.show()
````

## Q6: Write a query to find the total count of records for each city in a DataFrame, but only for cities where the count is greater than 100.

````python
# Import required functions
from pyspark.sql.functions import count, col

# Sample Data
data = [
    ("Delhi",),
    ("Delhi",),
    ("Mumbai",),
    ("Delhi",),
    ("Mumbai",),
    ("Chennai",)
]

columns = ["city"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

print("Original Data:")
df.show()

# Group by city and count records
city_count = df.groupBy("city") \
               .agg(count("*").alias("Total_Count"))

# Filter cities where count is greater than 100
result = city_count.filter(col("Total_Count") > 100)

print("Cities with Count Greater Than 100:")
result.show()
````

## Q7: How does the immutability of Spark DataFrames affect how you perform "data cleaning" steps like dropping columns or renaming them?


Spark DataFrames are **immutable**, which means their data cannot be changed after they are created.

During data-cleaning operations such as:

- Dropping columns
- Renaming columns
- Filtering rows
- Removing duplicates

Spark does not modify the original DataFrame. Instead, it creates a **new DataFrame** containing the requested changes.

This approach provides:

- Better fault tolerance
- Easier debugging
- Consistent data processing
- Safe parallel execution

## Example: Dropping a Column

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("ImmutabilityExample").getOrCreate()

# Sample Data
data = [
    (1, "Ali", 50000),
    (2, "Sara", 60000)
]

columns = ["id", "name", "salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Drop salary column
df_drop = df.drop("salary")

print("Original DataFrame:")
df.show()

print("New DataFrame After Dropping Column:")
df_drop.show()
```

## Example: Renaming a Column

```python
# Rename name column to employee_name
df_rename = df.withColumnRenamed("name", "employee_name")

print("DataFrame After Renaming Column:")
df_rename.show()
```

## Output

### Original DataFrame

```text
+---+----+------+
| id|name|salary|
+---+----+------+
|  1| Ali| 50000|
|  2|Sara| 60000|
+---+----+------+
```

### After Dropping Column

```text
+---+----+
| id|name|
+---+----+
|  1| Ali|
|  2|Sara|
+---+----+
```

### After Renaming Column

```text
+---+-------------+------+
| id|employee_name|salary|
+---+-------------+------+
|  1|          Ali| 50000|
|  2|         Sara| 60000|
+---+-------------+------+
```

## Conclusion

Because Spark DataFrames are immutable, every data-cleaning operation returns a new DataFrame instead of modifying the original one. Therefore, when dropping or renaming columns, you must store the result in a new DataFrame (or overwrite the existing variable) to use the cleaned data.

## Q8: Write a Spark command to filter a dataset for rows where the age is between 18 and 30 (inclusive) and the subscription is 'Premium'.

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Sample Data
data = [
    (18, "Premium"),
    (22, "Premium"),
    (25, "Basic"),
    (30, "Premium"),
    (35, "Premium")
]

columns = ["age", "subscription"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Filter condition
filtered_df = df.filter(
    (col("age").between(18, 30)) &
    (col("subscription") == "Premium")
)

filtered_df.show()
````

## Explanation

between(18, 30) → selects ages from 18 to 30 inclusive.

& → combines multiple conditions using AND.

subscription == "Premium" → filters only Premium subscribers.

## Q10: Write the code to revise a column named raw_timestamp by casting it to a TimestampType and renaming it to event_time.

````python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType


data = [
    (25, "Premium", "2026-06-23 10:00:00"),
    (30, "Basic", "2026-06-23 11:00:00")
]

columns = ["age", "subscription", "raw_timestamp"]

df = spark.createDataFrame(data, columns)

df_new = df.withColumn(
    "event_time",
    col("raw_timestamp").cast("timestamp")
).drop("raw_timestamp")

df_new.show()
````

Explanation

cast(TimestampType()) → converts string/raw value into proper timestamp format

withColumn("event_time", ...) → creates renamed column

drop("raw_timestamp") → removes old column after conversion

## Q11: Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?


## What is Shuffle in Spark?

Shuffle is the process in which Spark **redistributes data across different partitions/executors** when a transformation requires grouping, joining, or aggregation.

It happens when records that are initially spread across different partitions need to be brought together based on a key.

## Example of Shuffle (groupBy operation)

Suppose we have data distributed like this:

- Partition 1 → Delhi, Mumbai  
- Partition 2 → Delhi, Chennai  
- Partition 3 → Mumbai, Delhi  

When we perform:

```python
df.groupBy("city").count()
```

Spark must move all "Delhi" records into one place and all "Mumbai" records into another. This movement across partitions is called **Shuffle**.

## Why Shuffle Happens?

Because operations like:
- groupBy()
- join()
- distinct()
- orderBy()

need data from multiple partitions to be combined.

## Why is Shuffle a Wide Transformation?

A transformation is called **wide** when:

- Data needs to move between multiple partitions
- One stage depends on data from many other partitions
- Network communication between executors is required

### In Shuffle:
- Data is transferred across the network
- Multiple partitions are involved in producing output
- Spark creates a new stage in execution plan

## Conclusion

Shuffle is a costly process because it involves disk I/O and network communication. It is classified as a wide transformation because it requires data movement across multiple partitions before producing the final result.

### One-Line Explanation
Shuffle is a wide transformation because it redistributes data across partitions over the network to complete operations like grouping and joining.

## Q13: How do you use the .agg() function to calculate multiple statistics at once, such as the min, max, and mean of the price column?

The .agg() function in PySpark is used to perform multiple aggregate operations on one or more columns in a single query.

In this example, we calculate:

Minimum Price

Maximum Price

Average (Mean) Price

using the price column.

````python
# Import required functions
from pyspark.sql.functions import min, max, avg

# Sample Data
data = [
    (100,),
    (250,),
    (300,),
    (450,),
    (500,)
]

columns = ["price"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

print("Original Data:")
df.show()

# Calculate multiple statistics using .agg()
df.agg(
    min("price").alias("Minimum_Price"),
    max("price").alias("Maximum_Price"),
    avg("price").alias("Average_Price")
).show()
````

## Q14: In the context of cleaning a dataset, what is the risk of using inferSchema=true when your source data contains messy or inconsistent date formats?


When reading data in PySpark, `inferSchema=True` automatically tries to detect the correct data types of each column.

However, when the dataset contains **messy or inconsistent date formats**, this can lead to several problems.

## Example of messy date formats

```
2026-06-23
23/06/2026
June 23, 2026
2026.06.23
```

## Risks of using `inferSchema=True`

### 1. Incorrect Data Type Inference
Spark may fail to recognize all formats as dates and treat them as **strings** instead of `DateType` or `TimestampType`.

### 2. Null Values Creation
Unrecognized date formats may be converted into **null values**, leading to data loss.

### 3. Inconsistent Parsing
Some rows may be parsed correctly while others fail, causing **inconsistent column behavior**.

### 4. Data Quality Issues
Wrong or missing date values can break:
- time-based analysis  
- sorting  
- filtering  
- aggregations  

### 5. Performance Overhead
Schema inference adds extra computation while reading large datasets.

## Conclusion

Using `inferSchema=True` is risky when date formats are inconsistent because it can lead to incorrect types, null values, and unreliable analysis results.

## Best Practice

Always define schema manually when working with:
- dates
- timestamps
- messy or real-world datasets

### One-Line Explanation
`inferSchema=True` can misinterpret inconsistent date formats, leading to wrong types, null values, and poor data quality.

##Q15: Write a final processing pipeline that:
1.Filters out duplicates.

2.Fills null prices with 0.

3.Groups by store_id to calculate total revenue.

This pipeline performs the following operations:

Removes duplicate records.

Replaces null values in the price column with 0.

Groups the data by store_id.

Calculates the total revenue for each store using the sum() function.

````python
# Import required function
from pyspark.sql.functions import sum

# Sample Data
data = [
    (101, 500),
    (101, None),
    (102, 700),
    (103, 900),
    (102, 700),   # Duplicate
    (104, None)
]

columns = ["store_id", "price"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

print("Original Data:")
df.show()

# Step 1: Remove duplicate rows
df = df.dropDuplicates()

# Step 2: Fill null prices with 0
df = df.na.fill({"price": 0})

# Step 3: Group by store_id and calculate total revenue
result = df.groupBy("store_id") \
           .agg(sum("price").alias("Total_Revenue"))

print("Final Result:")
result.show()
````
